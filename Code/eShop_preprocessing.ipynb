# Transformation of the UCI "e-shop" dataset in CSV format

The main objective is to transform this UCI dataset into a CSV file with the following format <Id, Date/Order, Attribute1, Attribute2,...>.

Load the original dataset (CSV format):

```python
import pandas as pd
import csv

e_shop = pd.read_csv('..\Data\Raw\e-shop data and description\e-shop clothing 2008.csv', header=0, sep=";", parse_dates={'Date' : ['year','month','day']})#, infer_datetime_format=True)
e_shop.head(10)
```

Rename columns and change their order:

```python
e_shop = e_shop.rename(columns={'order':'Order', 'country':'Country', 'session ID':'Id', 'page 1 (main category)':'Main category', 'page 2 (clothing model)':'Clothing model', 'colour':'Colour', 'location':'Location', 'model photography':'Model photography', 'price':'Price', 'price 2':'Price higher avg', 'page':'Page nber'})

column_names = ['Id', 'Date', 'Order', 'Country', 'Main category', 'Clothing model', 'Colour', 'Location', 'Model photography', 'Price', 'Price higher avg', 'Page nber'] 
e_shop = e_shop.reindex(columns=column_names)

e_shop.head(10)
```

Delete the "Date" column because only the "Order" is used to generate sequences (date is always identical for a given purchase):

```python
e_shop = e_shop.drop('Date',1)
e_shop.head(10)
```

Save in CSV format the sequences of products displayed in all user sessions:


```python
e_shop.to_csv('../Data/CSV/e_shop.csv', index=False, sep=';')
```
